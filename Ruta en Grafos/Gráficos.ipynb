```python
import matplotlib.pyplot as plt

k = [1000, 3000, 5000, 7000, 9000]
tiempo_k = [2.60, 2.39, 2.59, 2.43, 2.40]

plt.plot(k, tiempo_k, marker='o')
plt.xlabel('k')
plt.ylabel('Tiempo de Ejecución (segundos)')
plt.title('n = 10000 / m = 10000')
plt.ylim(0, 5)

for i in range(len(k)):
    plt.text(k[i], tiempo_k[i]+0.1, str(tiempo_k[i]), ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt

n1 = [4000, 6000, 8000, 10000, 12000, 14000, 16000]
tiempo1 = [0.0156, 0.046, 0.0625, 0.1, 0.14, 0.25, 0.43]

plt.plot(n1, tiempo1, marker='o')
plt.xlabel('n')
plt.ylabel('Tiempo de Ejecución (segundos)')
plt.title('k = 2 / m = 10')
plt.ylim(0, 1)  # Establecer el rango del eje y entre 0 y 1

for i in range(len(n1)):
    plt.text(n1[i], tiempo1[i]+0.05, str(tiempo1[i]), ha='center', va='bottom')

plt.show()
```
