```python
import matplotlib.pyplot as plt
import numpy as np

# Crear figura con 1 fila y 2 columnas
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Grfico 1 PD
n1 = [1000, 10000, 100000, 1000000, 10000000, 100000000]
tiempo1 = [0, 0.015625, 0.0625, 0.609375, 4.90625, 50.4844]

axs[0].plot(n1, tiempo1, marker='o')
axs[0].set_xscale('log')    # Establecer escala logartmica en el eje x
axs[0].set_xlabel('Tamao Escalera (n)')
axs[0].set_ylabel('Tiempo de Ejecucin (segundos)')
axs[0].set_title('Base Potencia = 2 / Rotos = 100')
for i in range(len(n1)):
    axs[0].text(n1[i], tiempo1[i]+0.6, str(round(tiempo1[i],2)), ha='center', va='bottom', fontsize=8)


# Grfico 2 PD
n2 = [16384, 32768, 65536, 131072, 262144, 524288, 1048576, 2097152, 4194304]
tiempo2 = [0.21875, 0.4375, 1.07812, 2.20312, 4.5, 9.5, 20.0938, 41.9688, 87.0312]

axs[1].plot(n2, tiempo2, marker='o')  # Graficar los datos
axs[1].set_xscale('log', base=2)    # Establecer escala logartmica en el eje x
axs[1].set_xlabel('Tamao Escalera (n)')    # Etiqueta del eje x
axs[1].set_title('Base Potencia = 2 / Rotos = 6000')    # Ttulo del grfico
for i in range(len(n2)):
    axs[1].text(n2[i], tiempo2[i]+0.8, str(round(tiempo2[i],2)), ha='center', va='bottom', fontsize=8)

def power_of_2_labels(x, pos):
    power = int(np.log2(x))
    return r'$2^{{{}}}$'.format(power)

ticks = [2**i for i in range(14,23)]
labels = [power_of_2_labels(t, i) for i, t in enumerate(ticks)]
axs[1].set_xticks(ticks, labels)

plt.tight_layout()  # Ajustar el espaciado entre subtramas
plt.show()
```

- En el grafico se aprecia una curva exponencial, pero como el eje X esta en escala logaritmica, esta curva exponencial se transforma en una linea recta (Lo cual confirma que el tiempo de ejecucion es O(n) para un k constante)
- En una escala logartmica, una curva exponencial se transforma en una linea recta. (Esto se debe a que la escala logartmica comprime los valores grandes en el eje x, lo que suaviza el crecimiento exponencial y lo convierte en un crecimiento ms lineal en el grfico.)
- Se muestran dos ejemplos para verificar que este bien hecho el experimento


```python
# Grfico 3 PD

r2 = [10, 100, 1000, 10000]
t2 = [0.046875, 0.296875, 2.03125, 14.0625]

plt.plot(r2, t2, marker='o')
plt.xlabel('Cantidad de Escalones rotos (r)')
plt.xscale("log")
plt.ylabel('Tiempo de Ejecucin (segundos)')
plt.title('n = 1000000 / Base Potencia = 5')

for i in range(len(r2)):
    plt.text(r2[i], t2[i]+0.1, str(round(t2[i],2)), ha='center', va='bottom')

plt.show()
```

```python
# Grfico 1 FB
n3 = [115, 120, 125, 130, 135, 140]
tiempo3 = [0, 0.171875, 1.82812, 18.3906, 39.8906, 82.375]

plt.plot(n3, tiempo3, marker='o')
plt.xlabel('Tamao Escalera (n)')
plt.ylabel('Tiempo de Ejecucin (segundos)')
plt.title('Base Potencia = 2 / Rotos = 70')

for i in range(len(n3)):
    plt.text(n3[i], tiempo3[i]+0.8, str(tiempo3[i]), ha='center', va='bottom')

plt.show()
```

- El grfico esta en escala lineal, por lo tanto la curva exponencial se interpreta como tal.
- Lo que quiere decir que el algoritmo de fuerza bruta es de orden exponencial, muy poco eficiente.
- A medida que crece n, el tiempo de ejecucion aumenta exponencialmente


```python
# Grfico 2 FB
r1 = [10, 11, 12, 13, 14, 15, 16, 17, 18]
t1 = [81.5781, 42.3438, 25.2344, 10.2969, 6.801, 2.8125, 1.2656, 0.671, 0.4843]

plt.plot(r1, t1, marker='o')
plt.xlabel('Cantidad de Escalones rotos (r)')
plt.ylabel('Tiempo de Ejecucin (segundos)')
plt.title('n = 100 / Base Potencia = 5')

for i in range(len(r1)):
    plt.text(r1[i], t1[i]+0.8, str(round(t1[i],2)), ha='center', va='bottom')

plt.show()
```

- Tenamos que el tiempo de ejecucion era k^n
- Realizamos pruebas cambiando el valor de r para ver como afectaba, n y p se mantuvieron fijos
- Podemos apreciar que a medida que incrementamos el nmero de escalones rotos, el tiempo de ejecucion baja significativamente
- Como n y p son constantes, k igual es constante, por lo tanto la nica forma de bajar el tiempo de ejecucion exponencialmente es restar en el exponente
- O(k^(n-r)), esto debido a que en la funcion formasFB, se hacen n-r llamadas recursivas.
