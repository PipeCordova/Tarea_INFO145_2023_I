  <p> Se muestra un gráfico para el algoritmo de Programación Dinámica (PD), donde se ejecutó para valores constantes de n = 100000 y r = 10. Se varió el parámetro p (potencia), pero solo pudimos ejecutar hasta p = 3 debido a que al utilizar p = 4 se produjo overflow.</p>

```python
import matplotlib.pyplot as plt

# Datos de entrada
p_values = [2, 3]
tiempos = [0.01278, 0.006441]

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(p_values, tiempos, marker='o', linestyle='-', color='b')
plt.xlabel('p (base pot)')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.title(f'Tiempo de ejecución en función de p\n(n = 100000, r = 10)')
plt.xticks(p_values)
plt.grid(True)

# Etiquetas de los puntos de datos
for i, txt in enumerate(tiempos):
    plt.annotate(f'Tiempo: {txt:.6f}', (p_values[i], tiempos[i]), xytext=(5, -10), textcoords='offset points')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

n1 = [1000, 10000, 100000, 1000000, 10000000, 100000000]
tiempo1 = [0, 0.015625, 0.0625, 0.609375, 4.90625, 50.4844]

fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # Crear figura con 1 fila y 2 columnas

# Gráfico 1
axs[0].plot(n1, tiempo1, marker='o')
axs[0].set_xscale('log')    # Establecer escala logarítmica en el eje x
axs[0].set_xlabel('Tamaño escalera (n)')
axs[0].set_ylabel('Tiempo de Ejecución (segundos)')
axs[0].set_title('Base Potencia = 2 / Rotos = 100')
for i in range(len(n1)):
    axs[0].text(n1[i], tiempo1[i]+0.6, str(tiempo1[i]), ha='center', va='bottom', fontsize=8)

n2 = [1000, 10000, 100000, 1000000, 10000000, 100000000]
tiempo2 = [0, 0.015625, 0.046875, 0.59375, 5.21875, 48]

# Gráfico 2
axs[1].plot(n2, tiempo2, marker='o')
axs[1].set_xscale('log')    # Establecer escala logarítmica en el eje x
axs[1].set_xlabel('Tamaño escalera (n)')
axs[1].set_ylabel('Tiempo de Ejecución (segundos)')
axs[1].set_title('Base Potencia = 50 / Rotos = 50')
for i in range(len(n2)):
    axs[1].text(n2[i], tiempo2[i]+0.6, str(tiempo2[i]), ha='center', va='bottom', fontsize=8)

plt.tight_layout()  # Ajustar el espaciado entre subtramas
plt.show()
```

- En el grafico se aprecia una curva exponencial, pero como el eje X esta en escala logaritmica, esta curva exponencial se transforma en una linea recta (Lo cual confirma que el tiempo de ejecucion es O(n) para un k constante)
- En una escala logarítmica, una curva exponencial se transforma en una linea recta. (Esto se debe a que la escala logarítmica comprime los valores grandes en el eje x, lo que suaviza el crecimiento exponencial y lo convierte en un crecimiento más lineal en el gráfico.)
- Se muestran dos ejemplos para verificar que este bien hecho el experimento
